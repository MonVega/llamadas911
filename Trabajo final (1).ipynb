Se observa, que el tipo indicente_c4 es el que tiene una mayor cantidad de datos nulos, como representa el 6% del total, se obta por eliminarlos.

```python
#Elimar filas con NaN
llamadasC5_new=llamadasC5.dropna(axis=0)
```

```python
#Verificar que ya no hay datos nulos
llamadasC5_new.isnull().sum()
```

### Tipo de datos
Observamos en la siguiente salida, que hay variables que no tienen el tipo de dato adecuado, como lo es fecha_creacion,hora_creacion, fecha_cierre y hora_cierre; por lo que se realizara el cambio adeacuado.

```python
llamadasC5_new.info()
```

```python
#Creacion de una nueva variable que contine la fecha y hora para creacion y cierre
llamadasC5_new['datetime_creacion']=llamadasC5_new['fecha_creacion']+' '+llamadasC5_new['hora_creacion']
llamadasC5_new['datetime_cierre']=llamadasC5_new['fecha_cierre']+' '+llamadasC5_new['hora_cierre']
```

```python
#Cambiamos el tipo de dato a datetime 
llamadasC5_new['datetime_creacion']=pd.to_datetime(llamadasC5_new['datetime_creacion'],format='%Y-%m-%d %H:%M:%S')
llamadasC5_new['datetime_cierre']=pd.to_datetime(llamadasC5_new['datetime_cierre'],format='%Y-%m-%d %H:%M:%S')
```

Dado que las columas creadas anteriormente contienen la información necesaria, se eliminan columnas que representan duplicidad( fecha-hora de creacion y fecha-hora cierre).

```python
#llamadasC5_new[llamadasC5_new['datetime_creacion'].dt.time > datetime.time(23,0)]
#llamadasC5_new['datetime_creacion'].dt.year
llamadasC5_new.drop(['anio_creacion','mes_creacion','fecha_creacion','hora_creacion','anio_cierre','mes_cierre','fecha_cierre','hora_cierre'], axis=1,inplace=True)
llamadasC5_new
```

```python
#Verificamos que el tipo de dato es adecuado 
llamadasC5_new.info()
```

### Errores en variables categoricas
Lo siguientes es verificar que no haya errores en las variables categoricas.
El primero es el codigo cierre, en este caso se observa que hay categorias que representan lo mismo pero tienen un valor diferente, en este caso reemplazamos con el valor con el adecuado.

```python
llamadasC5_new['codigo_cierre'].unique()
```

```python
#Reemplazo del el nombre de categorias
llamadasC5_new['codigo_cierre'].replace({'A':'Afirmativo','D':'Duplicado','I':'Informativo','N':'Negativo'},inplace=True) #inplace= remplazarlo en el mismo df
```

La variable clas_con_f_alarma, presenta un error ya que FALTA CÃ\x8dVICA representa lo mismo que FALTA CÍVICA, por lo cual se corrige.

```python
llamadasC5_new['clas_con_f_alarma'].unique()
```

```python
#Se hace el reemplazo
llamadasC5_new['clas_con_f_alarma'].replace({'FALTA CÃ\x8dVICA':'FALTA CÍVICA'},inplace=True)
#inplace= remplazarlo en el mismo df
```

se hace el reemplazo con la categoria Abandono que no contiene espacios en blanco.

```python
llamadasC5_new['categoria_incidente_c4'].replace({'Abandono ':'Abandono'},inplace=True)
```
